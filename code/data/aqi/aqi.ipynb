```python
from google.colab import drive
import pandas as pd
import os

drive.mount('/content/drive')
```

```python
folder_path = '/content/drive/MyDrive/CPSC 8810 Bio/Final Project/code/data/aqi'

ne = "/northeast"
se = "/southeast"
nw = "/northwest"
sw = "/southwest"

regions = {ne, se, nw, sw}
#mean_aqi_values = []

for region in regions:
    mean_aqi_values = []
    std_aqi_values = []
    csv_files = [f for f in os.listdir(folder_path + str(region)) if f.endswith('.csv')]

    for file in csv_files:
        file_path = os.path.join(folder_path + str(region), file)
        df = pd.read_csv(file_path)
        if 'Overall AQI Value' in df.columns:
            mean_aqi = df['Overall AQI Value'].mean()
            std_aqi = df['Overall AQI Value'].std()

            mean_aqi_values.append(mean_aqi)
            std_aqi_values.append(std_aqi)
        else:
            print(f"'Overall AQI Value' Column not found in {file}")
            break
    if mean_aqi_values:
        overall_mean_aqi = sum(mean_aqi_values) / len(mean_aqi_values)
        overall_std_aqi = pd.Series(mean_aqi_values).std()
    else:
        overall_mean_aqi = None
        overall_std_aqi = None

    print("--------", region, "--------")
    print("Mean AQI values per file:", mean_aqi_values)
    print("Overall mean AQI value:", overall_mean_aqi)
    print("Standard Deviation of AQI values per file:", std_aqi_values)
    print("Overall standard deviation of AQI values:", overall_std_aqi)
```
